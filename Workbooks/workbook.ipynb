# Data Cleaning:

```python
google_play_df.info
```

```python
google_play_df.isna().sum()
```

```python
clean_df = google_play_df.dropna(axis = 1)
clean_df.isna().sum()
```

```python
clean_df.head()
```

```python
clean_df['Installs'].value_counts()
```

```python
clean_df.loc[clean_df['Installs'] == '0']
```

```python
clean_df.drop(clean_df.loc[clean_df['Installs'] == '0'].index, inplace = True)
```

```python
clean_df['Installs'].value_counts()
```

```python
duplicates = clean_df[clean_df.duplicated(subset = 'App')]
print(len(duplicates))
duplicates.tail(15)
```

```python
len(clean_df['App'].unique())
```

```python
clean_df.drop_duplicates(subset = ['App'], inplace = True)
```

```python
clean_df.info
```

```python
clean_df.to_csv(index = False)
```
